```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

Steps:
- Load in the selected playlist data
    - This will be from the other functions
- With this data we will create the vibe
- Get Data from subreddit
    - This is already called in from MongoDB
- Preprocess data
- Fit cosine similarity
- Export top 'n' dataframe to be displayed

```python
def generate_similar_songs(reddit_songs, vibe):
    '''
    reddit_songs (pandas DataFrame): songs in the selected subreddit
    vibe (numpy arrary): array of vibe metric
    
    Output:
    Pandas DataFrame with 'cosine_sim' (cosine similarity metric) generated
    '''
    reddit_array = reddit_songs.to_numpy()
    cos_sim = {'cosine_sim' : []}
    for track in reddit_array:
        similarity = cosine_similarity(np.array([track]), vibe.T.to_numpy())
        cos_sim['cosine_sim'].append(similarity)
    
    return cos_sim
```
